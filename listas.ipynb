```python
#3.organizar una lista con el metodo burbuja
def burbuja(lista):
    for n in range(len(lista) - 1, 0, -1):
        for i in range(n):
            if lista[i] > lista[ i + 1]:
                temp = lista[i]
                lista[i]= lista[i + 1]
                lista[i + 1]= temp

numeros = [6, 4, 8, 5, 1, 3, 7, 9, 2]
print(numeros)


burbuja(numeros)
print(numeros)
       

    
    

```
